```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
# Generate a large dataset with missing values
np.random.seed(42)
num_rows = 10000
data = {
    'Feature1': np.random.normal(0, 1, num_rows),
    'Feature2': np.random.choice([1, 2, 3, 4, np.nan], num_rows),
    'Feature3': np.random.choice(['A', 'B', 'C', np.nan], num_rows)
}
```

```python
df = pd.DataFrame(data)
df
```

```python
# Mean imputation
mean_imputed = df.copy()
mean_imputed['Feature2'].fillna(df['Feature2'].mean(), inplace=True)
```

```python
# Median imputation
median_imputed = df.copy()
median_imputed['Feature2'].fillna(df['Feature2'].median(), inplace=True)
```

```python
# Mode imputation
mode_imputed = df.copy()
mode_imputed['Feature3'].fillna(df['Feature3'].mode()[0], inplace=True)
```
