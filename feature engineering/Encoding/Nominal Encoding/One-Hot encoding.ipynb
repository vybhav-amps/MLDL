```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('homeprices.csv')
data
```

### since the above data has string data in data['town'] we need to convert it into 1's and 0's to make the machine understandable.

```python
# Perform one-hot encoding on the 'town' column
one_hot_encoded = pd.get_dummies(data['town'])
```

```python
# Concatenate the original data with the one-hot encoded data
encoded_data = pd.concat([data, one_hot_encoded], axis=1)
```
