```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('segmentation data.csv')
df.head(10)
```

```python
df.drop('ID', axis=1, inplace=True)
df.head()
```

```python
df.isna().sum()
```

```python
df.hist(figsize =(10,10))
plt.show()
```

```python
y_predict=[]

for i in range(2,6):

    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42, n_init=10)
    y_predict.append(kmeans.fit_predict(df))

for i in range(len(y_predict)):

    print(y_predict[i])
```

```python
df['2_Clusters'] = y_predict[0]
df['3_Clusters'] = y_predict[1]
df['4_Clusters'] = y_predict[2]
df['5_Clusters'] = y_predict[3]
```

```python
plt.figure(figsize=(15,6))
plt.scatter(df["Age"],df["Income"])

plt.xlabel('Age')
plt.ylabel('Income')
plt.show()
```

with 2 clusters

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Age'], df['Income'], c=df['2_Clusters'], cmap='viridis')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('K-means Clustering')
plt.show()
```
